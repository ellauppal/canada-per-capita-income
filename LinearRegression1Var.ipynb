```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
# Read csv with data
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
# Create scatter plot with data
plt.scatter(df.year,df.capita, c='blue', edgecolor='black', alpha=0.75)
plt.title('Canada\'s Per Capita Income')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
```

```python
# Test train split for supervised training
x_train, x_test, y_train, y_test = train_test_split(df.year, df.capita, test_size=0.2)
```

```python
# Test train split visulization
plt.scatter(x_train, y_train, label='Training Data', c='red', edgecolor='black', alpha=0.75)
plt.scatter(x_test, y_test, label='Testing Data', c='green', edgecolor='black', alpha=0.75)
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.legend()
plt.show()
```

```python
# Create linear model and train it

# Create linear regression object
LR = linear_model.LinearRegression()

# Pass training data
LR.fit(x_train.values.reshape(-1,1), y_train.values)
```

```python
# Use model to predict output(ŷ) for TEST data input(x)
prediction = LR.predict(x_test.values.reshape(-1,1))

# Plot prediction line (x,ŷ)
plt.plot(x_test, prediction, label='Linear Regression', c='red')

# Plot actual test data (x,y)
plt.scatter(x_test,y_test, c='green', edgecolor='black', alpha=0.75, label='Actual Test Data')

plt.legend()
plt.show()
```

```python
# Predict per capita income for year 2020
LR.predict([[2020]])[0]
```

```python
# Score this model
LR.score(x_test.values.reshape(-1,1),y_test.values)
```
